# Seaborn Basics

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

Let's discuss some plots that allow us to visualize the data set. These plots are:

* Scatter Plot
* Line Plot
* Bar Graph
* Histogram
* Heatmap



## Imports

```python
import seaborn as sns
%matplotlib inline
```

## Data
Seaborn comes with built-in data sets!

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## Scatter Plot

Scatter plots are the graphs that present the relationship between two variables in a data-set.

```python
# Create scatter plot using seaborn
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

Assigning a variable to hue will map its levels to the color of the points:

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
```

A numeric variable can also be assigned to size to apply a semantic mapping to the areas of the points:

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", size="size")
```

## Line Plot

 A line graph, also known as a line chart or a line plot, is commonly drawn to show information that changes over time. You can plot it by using several points linked by straight lines. It comprises two axes called the x-axis and the y-axis.

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
may_flights = flights[flights["month"] == 'May'] #Filtering the dataset to get the records for the month of May
sns.lineplot(data=may_flights, x="year", y="passengers")
```

Assign a grouping semantic (hue, size, or style) to plot separate lines

```python
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
```

## Bar Plot

A bar plot represents an estimate of central tendency for a numeric variable with the height of each rectangle

```python
df = sns.load_dataset("penguins")
df.head()
```

Group by a categorical varaible and plot aggregated values

```python
sns.barplot(data=df, x="island", y="body_mass_g")
```

Add a second layer of grouping:

```python
sns.barplot(data=df, x="island", y="body_mass_g", hue="sex")
```

## Histogram

A histogram is a classic visualization tool that represents the distribution of one or more variables by counting the number of observations that fall within discrete bins.

```python
sns.histplot(data=df, x="flipper_length_mm")
```

Flip the plot by assigning the data variable to the y axis:

```python
sns.histplot(data=df, y="flipper_length_mm")
```

Check how well the histogram represents the data by specifying a different bin width:

```python
sns.histplot(data=df, x="flipper_length_mm", binwidth=10)
```

```python
sns.histplot(data=df, x="flipper_length_mm", binwidth=3)
```

You can draw multiple histograms from a long-form dataset with hue mapping:

```python
sns.histplot(data=df, x="flipper_length_mm", hue="species")
```

## Heatmap

Plot rectangular data as a color-encoded matrix.

```python
corr_df = df.corr()
corr_df
```

Pass a DataFrame to plot with indices as row/column labels:

```python
sns.heatmap(corr_df)
```

Use annot to represent the cell values with text:

```python
sns.heatmap(corr_df, annot=True)
```

Select a different colormap by name:

```python
sns.heatmap(corr_df, annot=True, cmap="crest")
```
